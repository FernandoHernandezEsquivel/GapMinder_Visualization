```python
chart = sns.histplot(gapminder_df.esperanza_vida, color='#98ff00', binwidth=10)
plt.title('Histograma de la esperanza de vida')
# Etiquetas del eje de las x's
plt.xlabel('Esperanza de vida')
# Etiquetas del eje de las y's
plt.ylabel('Frecuencia')
# Modificando los ejes
plt.ylim(0, None)
plt.xlim(0, None)
# Agregar la leyenda
plt.legend(['Esperanza de vida'])
# For loop para las anotaciones
for p in chart.patches:
            chart.annotate("%.0f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
            textcoords='offset points')
```

## Correlaciones lineales

Primero se calculan las correlaciones

```python
corr = gapminder_df.corr()
```

A partir de las correlaciones calculadas se grafican estas por medio de la funci√≥n heatmap

```python
sns.heatmap(corr,annot=True);
```

### Cambiando la paleta de color

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu");
```

```python
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
  f, ax = plt.subplots(figsize=(8, 8))
  sns.heatmap(corr,annot=True,mask=mask,vmax=.3,square=True)
```

```python
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
  f, ax = plt.subplots(figsize=(8, 8))
  sns.heatmap(corr,annot=True,cmap="YlGnBu",mask=mask,vmax=.3,square=True)
```
